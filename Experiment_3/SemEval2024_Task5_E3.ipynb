```python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0,
    )
    return response.choices[0].message["content"]
```

```python
def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    print(m.name)
```

```python
model = genai.GenerativeModel('gemini-pro')
```

```python
# results = []

for index, row in df.iterrows():

    if index < 4:
        continue

    time.sleep(2)
    # if index % 3 == 0 and index != 0:
    #   time.sleep(62)

    prompt = (
        "I will provide a question, an answer, and an explanation. "
        "Your task is to determine if the answer is correct based on the explanation provided. "
        "After reading the explanation, please respond with 'yes' if the answer is correct, "
        "or 'no' if it is incorrect. \n\n"
        f"Question: {row['question']}\n"
        f"Proposed Answer: {row['answer']}\n"
        f"Explanation: {row['explanation']}\n\n"
        "Is the proposed answer correct based on the explanation? (yes or no)\n"
        "Please provide your detailed reason for your choice.\n"
        "Then, reevaluate and check whether the selected answer is logical or not. "
        "Please use the following format:\n"
        "<selected_answer>: yes/no\n"
        "<reason>: your reason for the initial choice\n"
        "<reason for logical check>: your reason for reevaluation"
    )

    print(index)
    response = model.generate_content(prompt).text
    print(response)
    result = 1 if 'yes' in response.lower() else 0
    print(result)
    print("*********************************")
    results.append({'index': index, 'prediction': result})
```

```python
results_df = pd.DataFrame(results)
results_df.to_csv('outputs_2.csv', index=False)
```
