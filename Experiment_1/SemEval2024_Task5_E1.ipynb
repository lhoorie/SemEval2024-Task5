# **Longformer**

```python
PATH_TO_SAVE = "./checkpoint_1"
PATH_FOR_READ = './checkpoint_1'
```

```python
tokenizer = LongformerTokenizer.from_pretrained('allenai/longformer-base-4096')
model = LongformerForSequenceClassification.from_pretrained('allenai/longformer-base-4096', ignore_mismatched_sizes=True)
```

```python
train_df = pd.read_csv('/content/dataset_semeval24_traindev/new_train.csv')
val_df = pd.read_csv('/content/dataset_semeval24_traindev/new_dev.csv')
```

```python
def get_tokenizer_output(tokenizer, text, attr):
    return np.array(tokenizer(text, padding='max_length', truncation=True, return_tensors="pt")[attr])

class LegalDataset(Dataset):
    def __init__(self, df, tokenizer, classes):
        super(LegalDataset, self).__init__()
        self.df = df
        self.tokenizer = tokenizer
        self.labels = []
        self.label2idx = {c:i for i, c in enumerate(classes)}
        self.idx2label = {i:c for i, c in enumerate(classes)}
        self.encodings = {}
        self.encodings['input_ids']=[]
        self.encodings['attention_mask']=[]

        for index, row in df.iterrows():
            text = row['question'] + row['answer'] + row['explanation']
            self.encodings['input_ids'].append(get_tokenizer_output(self.tokenizer, text, 'input_ids'))
            self.encodings['attention_mask'].append(get_tokenizer_output(self.tokenizer, text, 'attention_mask'))
            self.labels.append(row['label'])
        self.encodings['input_ids'] = np.array(self.encodings['input_ids'])
        self.encodings['attention_mask'] = np.array(self.encodings['attention_mask'])

        self.labels = np.array(self.labels).reshape(-1, 1)


    def __len__(self):
        return len(self.labels)


    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item
```

```python
classes = np.array([0, 1])

train_dataset = LegalDataset(train_df, tokenizer, classes)
val_dataset = LegalDataset(val_df, tokenizer, classes)


print("ٔNumber of samples in train split", len(train_dataset))
print("ٔNumber of samples in val split", len(val_dataset))

train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=1, shuffle=True, num_workers=4)
val_dataloader = torch.utils.data.DataLoader(val_dataset, batch_size=1, shuffle=False, num_workers=4)
```

```python
class FocalLoss(nn.Module):
    def __init__(self, gamma=2, weight=None, ignore_index=-100):
        super().__init__()
        self.gamma = gamma
        self.weight = weight
        self.ignore_index = ignore_index

    def forward(self, input, target):
        logit = F.log_softmax(input, dim=1)
        pt = torch.exp(logit)
        logit = (1 - pt)**self.gamma * logit
        loss = F.nll_loss(
            logit, target, self.weight, ignore_index=self.ignore_index)
        return loss
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
num_classes = len(classes)
model = model.to(device)

criterion = FocalLoss()

optimizer = optim.AdamW(model.parameters(), lr=5e-5)
num_epochs = 3
best_val_loss = float('inf')

for epoch in range(num_epochs):
    model.train()
    total_loss = 0.0
    total_correct = 0
    total_samples = 0
    for train_batch in tqdm(train_dataloader):
        input_ids = train_batch['input_ids'].squeeze(1).to(device)
        attention_mask = train_batch['attention_mask'].squeeze(1).to(device)
        train_label = train_batch['labels'].squeeze(1).to(device)

        outputs = model(input_ids, attention_mask)
        logits = outputs.logits

        _, preds = torch.max(logits, dim=1)
        loss = criterion(logits, train_label)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        total_correct += (preds == train_label).sum().item()
        total_samples += train_label.size(0)
        optimizer.zero_grad()
    print(f"Train, Epoch {epoch + 1} : Loss = {total_loss / len(train_dataloader):.4f}, Accuracy = {total_correct / total_samples:.4f}")


    model.eval()
    correct = 0
    total = 0
    total_loss = 0.0
    total_correct = 0
    total_samples = 0
    val_preds = []
    val_labels = []


    with torch.no_grad():
        for val_batch in tqdm(val_dataloader):
            input_ids = val_batch['input_ids'].squeeze(1).to(device)
            attention_mask = val_batch['attention_mask'].squeeze(1).to(device)
            val_label = val_batch['labels'].squeeze(1).to(device)

            outputs = model(input_ids, attention_mask)
            logits = outputs.logits
            _, preds = torch.max(logits, dim=1)

            val_preds.append(preds)
            val_labels.append(val_label)
            loss = criterion(logits, val_label)
            total_loss += loss.item()
            total_correct += (preds == val_label).sum().item()
            total_samples += val_label.size(0)

        val_preds = torch.cat(val_preds)
        val_labels = torch.cat(val_labels)
        f1 = f1_score(val_labels.cpu(), val_preds.cpu(), average='macro')

        if total_loss > best_val_loss:
          torch.save(model.state_dict(), f'{PATH_TO_SAVE}')
          best_val_loss = total_loss

        print(f"Evaluation, Epoch {epoch + 1} : Loss = {total_loss / len(val_dataloader):.4f}, Accuracy = {total_correct / total_samples:.4f}, F1 Score: {f1}")
```

```python
model.load_state_dict(torch.load(f'{PATH_TO_SAVE}'))
```
